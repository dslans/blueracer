```python
logfolder = 'logs/DL-racingline-speed-progress/logs/training'
logfile = 'training-20210208174751-heyBXCL2QP6y5s-b0BeUpQ-robomaker.log'
training = la.load_data(f'{logfolder}/{logfile}')
training_df = la.convert_to_pandas(training)
```

```python
df = training_df[training_df.iteration==1]
df.tail()
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf

model = smf.ols(formula='reward ~ steps + progress + steps:progress + steer + throttle', data=training_df).fit()

# y = df.reward
# x = df[['steps','steer','throttle','progress']]
# model = sm.OLS(y, x).fit()
model.summary()
```

```python
model.params[1:]
```

```python
pd.DataFrame(model.params[1:]).T
```

```python
iterations = training_df.iteration.unique()
param_df = pd.DataFrame()
for i in iterations:
    df = training_df[training_df.iteration==i]
    model = smf.ols(formula='reward ~ steps + progress + steps:progress + steer + throttle', data=df).fit()
    params = pd.DataFrame(model.params[1:]).T
    param_df = pd.concat([param_df, params])
param_df = param_df.reset_index(drop=True)
param_df
```

```python
# Plot effect of parameters on reward size
ax = param_df.plot(figsize=(10,6))
```

```python
def plot_tracknumbers(trackpoints, show=True):
    import matplotlib.pyplot as plt
    for i in range(len(trackpoints)):
        plt.scatter(trackpoints[i][0], trackpoints[i][1], marker=f"${str(i).zfill(2)}$", c='black', s=200, edgecolors='none')
        plt.scatter(trackpoints[i][2], trackpoints[i][3], marker=f"${str(i).zfill(2)}$", c='black', s=200, edgecolors='none')
        plt.scatter(trackpoints[i][4], trackpoints[i][5], marker=f"${str(i).zfill(2)}$", c='black', s=200, edgecolors='none')
    if show:
        plt.show()
```

```python
plt.figure(figsize=(16,10))
plot_tracknumbers(track)
```

```python
def plot_tracknumbers(trackpoints, show=True):
    import matplotlib.pyplot as plt
    for i in range(len(trackpoints)):
        plt.scatter(trackpoints[i][0], trackpoints[i][1], c='blue', s=8)
        plt.scatter(trackpoints[i][2], trackpoints[i][3], c='red', s=8)
        plt.scatter(trackpoints[i][4], trackpoints[i][5], c='red', s=8)
        
        plt.annotate(f'{i}', # this is the text
             (trackpoints[i][0],trackpoints[i][1]), # this is the point to label
             textcoords="offset points", # how to position the text
             xytext=(0,10), # distance from text to points (x,y)
             ha='center') # horizontal alignment can be left, right or center
        
        plt.annotate(f'{i}', # this is the text
             (trackpoints[i][2],trackpoints[i][3]), # this is the point to label
             textcoords="offset points", # how to position the text
             xytext=(0,10), # distance from text to points (x,y)
             ha='center') # horizontal alignment can be left, right or center
        
        plt.annotate(f'{i}', # this is the text
             (trackpoints[i][4],trackpoints[i][5]), # this is the point to label
             textcoords="offset points", # how to position the text
             xytext=(0,10), # distance from text to points (x,y)
             ha='center') # horizontal alignment can be left, right or center
    if show:
        plt.show()


```

```python
plt.figure(figsize=(16,10))
plot_tracknumbers(track)
```
